a. If the weekly sales are affected by the unemployment rate, if yes - which stores are suffering the most? 

```python
correlation = Walmart_data['Weekly_Sales'].corr(Walmart_data['Unemployment'])
correlation
```

```python
# Visualize the relationship with a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Unemployment', y='Weekly_Sales', data=Walmart_data.head(200),color ='Brown')
plt.title(f'Correlation: {correlation:.2f}')
plt.xlabel('Unemployment Rate')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
# Identify stores with the highest negative correlation

stores_most_affected = Walmart_data.groupby('Store')['Weekly_Sales'].corr(Walmart_data['Unemployment']).idxmin()

print(f"The store(s) suffering the most due to unemployment: {stores_most_affected}")
```

The groupby operation is used to calculate the correlation for each store, and idxmin() is used to find the store with the highest negative correlation.

Note: A negative correlation suggests that as the unemployment rate increases, weekly sales tend to decrease.

```python
# Calculate the first and third quartiles
Q1 = Walmart_data['Weekly_Sales'].quantile(0.25)
Q3 = Walmart_data['Weekly_Sales'].quantile(0.75)
```

```python
# Calculate the IQR (Interquartile Range)
IQR = Q3 - Q1
IQR
```

```python
# Define upper and lower bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Filter out rows with values beyond the bounds

Good_data = Walmart_data[(Walmart_data['Weekly_Sales'] >= lower_bound) & (Walmart_data['Weekly_Sales'] <= upper_bound)]
Good_data
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(store_1['Weekly_Sales'])
print(result)
```

For Store_1 , p-Value is very high (P_Value >0.05 so we fail reject the null hypothesis ,it conferms that this data is not stationary with help of Time series diffrencing or Rolling mean we will try to make data stationary...

```python
from matplotlib import pyplot as plt

mean_data = store_1.rolling(window=52).mean()
std_data = store_1.rolling(window=52).std()

plt.plot(store_1, color = 'blue', label= 'Original')
plt.plot(mean_data, color = 'red', label= 'Rolling mean')
plt.plot(std_data, color = 'green', label= 'Rolling std')
plt.legend()
plt.show()
```

```python
## To make the data stationary
log_trans = np.log(store_1)
print(log_trans)
```

```python
log_trans.plot()
```

```python
result = adfuller(log_trans['Weekly_Sales'])
print(result)
```

```python
mean_log = log_trans.rolling(window=52).mean()
new_data = log_trans - mean_log
new_data = new_data.dropna()
```

```python
result = adfuller(new_data['Weekly_Sales'])
print(result)
```

```python
new_data.plot()
```

```python
from matplotlib import pyplot as plt

mean_data = new_data.rolling(window=52).mean()
std_data = new_data.rolling(window=52).std()

plt.plot(new_data, color = 'blue', label= 'Original')
plt.plot(mean_data, color = 'red', label= 'Rolling mean')
plt.plot(std_data, color = 'green', label= 'Rolling std')
plt.legend()
plt.show()
```

```python
## Seasonal decompose
from statsmodels.tsa.seasonal import seasonal_decompose

dec_result = seasonal_decompose(new_data['Weekly_Sales'], model='additive',period=7)

dec_result.plot()
```

```python
dec_result.seasonal
```

```python
# Train-test split
train_size = int(len(new_data) * 0.8)
train, test = new_data[:train_size], new_data[train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA

train = new_data.iloc[:105]['Weekly_Sales']
test = new_data.iloc[105:]['Weekly_Sales']
```

```python
train 
```

```python
test
```

```python
model = ARIMA(train, order=(1,0,2))
model_fit = model.fit()
print(model_fit.summary())
```

```python
## SARIMA

from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(train, order=(1,0,2), seasonal_order=(1,0,2,7))
```

```python
model_fit = model.fit()
```

```python
new_data['predict'] = model_fit.predict(start = 75, end=len(train)+len(test)-1)
new_data[['Weekly_Sales', 'predict']].plot()
```
